```python
error = np.loadtxt('/home/yuchen/ASRL/temp/lidar/aeva/boreas-2022-04-05-23-35/boreas-2022-04-05-23-35.no_rv/rv_error.log')
error1 = np.loadtxt('/home/yuchen/ASRL/temp/lidar/aeva/boreas-2022-04-05-23-35/boreas-2022-04-05-23-35.l2_cov1.0/rv_error.log')
error.shape
```

```python
fig, ax = plt.subplots(1, 1, figsize=(12, 6))
x = np.arange(error.shape[0])
# no rv
ax.plot(x, error[:, 0], 'r-', label='reference')
ax.plot(x, 3 * np.sqrt(error[:, 1]), 'r--', alpha=0.5)
ax.plot(x, -3 * np.sqrt(error[:, 1]), 'r--', alpha=0.5)
ax.fill_between(x, -3 * np.sqrt(error[:, 1]), 3 * np.sqrt(error[:, 1]), color='r', alpha=0.1)

ax.set_xlabel('frame')
ax.set_ylabel('radial velocity error [$m/s$]')
ax.legend()
ax.set_ylim([-5.0, 5.0])
ax.set_xlim([0, error.shape[0]])
ax.grid(alpha=0.5)

error = error1

fig, ax = plt.subplots(1, 1, figsize=(12, 6))
x = np.arange(error.shape[0])
# no rv
ax.plot(x, error[:, 0], 'r-', label='Geman-McClure (k=1.0)')
ax.plot(x, 3 * np.sqrt(error[:, 1]), 'r--', alpha=0.5)
ax.plot(x, -3 * np.sqrt(error[:, 1]), 'r--', alpha=0.5)
ax.fill_between(x, -3 * np.sqrt(error[:, 1]), 3 * np.sqrt(error[:, 1]), color='r', alpha=0.1)

ax.set_xlabel('frame')
ax.set_ylabel('radial velocity error [$m/s$]')
ax.legend()
ax.set_ylim([-5.0, 5.0])
ax.set_xlim([0, error.shape[0]])
ax.grid(alpha=0.5)
```
